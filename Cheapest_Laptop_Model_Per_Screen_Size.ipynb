## Import libraries

```python
import pandas as pd
import numpy as np
from numpy import random
```

```python
laptop_data = pd.read_csv('/datasets/LaptopsPrice.csv')
```

```python
laptop_data.head()
```

### Adding new column to specify if expensive

```python
is_expensive = laptop_data["Final Price"] > 2000
laptop_data["Expensive"] = is_expensive
laptop_data
```

#### Feature encoding

```python
laptop_data['Expensive'] = laptop_data['Expensive'].astype(int)
laptop_data.head()
```

```python
apple_hp_laptops = laptop_data[(laptop_data["Brand"] == "Apple") | (laptop_data["Brand"] == "HP")]
```

```python
apple_hp_laptops.value_counts('Brand')
```

### Filter out cheap (final price<=1200) laptops of Apple and HP

```python
apple_hp_laptops
```

#### Adding new column based on the condition/status of laptop

```python
conditions = [apple_hp_laptops["Status"] == "New", apple_hp_laptops["Status"] == "Refurbished"]
values = ["NE", "RE"]

laptops_condition = np.select(conditions, values, "NA")
```

```python
apple_hp_laptops["Condition"] = laptops_condition
apple_hp_laptops
```

```python
apple_hp_laptops.describe()
```

```python
conditions = [(apple_hp_laptops["Screen"] > 10) & (apple_hp_laptops["Screen"] < 14),
              (apple_hp_laptops["Screen"] >= 14) & (apple_hp_laptops["Screen"] < 17),
              (apple_hp_laptops["Screen"] >=17)]
values = ["Small", "Medium", "Large"]

laptop_sizes = np.select(conditions, values, "N/A")
apple_hp_laptops['Overall_size'] = laptop_sizes

apple_hp_laptops.value_counts("Overall_size")
```

```python
new_indices = np.arange(0, len(apple_hp_laptops), 1, dtype=int)
apple_hp_laptops["new_indices"] = new_indices

apple_hp_laptops.set_index("new_indices", inplace=True)
```

```python
del apple_hp_laptops["Status"]
del apple_hp_laptops["Laptop"]
del apple_hp_laptops["Screen"]
```

```python
apple_hp_laptops.head()
```

### Filtering out HP laptops with GPU GTX 1650 or RTX 3050: using isin()

```python
apple_hp_laptops
```

```python
first_cond = apple_hp_laptops["Brand"] == "HP"
second_cond = apple_hp_laptops["GPU"].isin(["GTX 1650", "RTX 3050"])
final_condition = first_cond & second_cond

hp_laptops = apple_hp_laptops[final_condition]
hp_laptops = pd.DataFrame(hp_laptops[["Brand", "Model", "Expensive", "Condition", "Overall_size", "Final Price"]])
```

### Finding the cheapest one for each size HP laptop

#### Sorting in ascending order based on Final Price

```python
hp_laptops.sort_values("Final Price", inplace=True)
hp_laptops
```

#### Removing duplicates

```python
hp_laptops = hp_laptops[~hp_laptops.duplicated(["Model", "Overall_size"], keep="first")]
```
